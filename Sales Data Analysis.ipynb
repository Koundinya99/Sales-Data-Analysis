# Data Analysis


### 1.What is the overall sales trend?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
#Getting month-year
df['month_year']=df['order_date'].apply(lambda x:x.strftime('%y-%m'))
```

```python
df['month_year']
```

```python
df.groupby(df['month_year']).sum()['sales']
```

```python
#Groupby monthly sales
df_trend=df.groupby(df['month_year']).sum()['sales'].reset_index()
```

```python
df_trend
```

```python
#SALES TREND
plt.figure(figsize=(20,9))
plt.xticks(rotation='vertical')
plt.plot(df_trend['month_year'],df_trend['sales'])
plt.show()
```

### 2.What are the top 10 products?

```python
prod_name=pd.DataFrame(df.groupby('product_name').sum()['sales'])
prod_name
```

```python
prod_name.sort_values('sales',ascending=False)
```

```python
prod_name.sort_values('sales',ascending=False).head(10)
```

```python
quan=pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
quan.sort_values('quantity')
```

```python
#TOP 10 PRODUCTS SOLD
quan.sort_values('quantity',ascending=False).head(10)
```

### 3.Which is the most profitable category and sub category?

```python
profit=pd.DataFrame(df.groupby(['category','sub_category']).sum()['profit'])
```

```python
#MOST PROFITABLE CATEGORY/SUB-CATEGORY
profit.sort_values(['category','sub_category'],ascending=False)
```
